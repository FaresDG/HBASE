# ROW_KEY IMPLEMENTATION

```python





def implement_rowkey(row):
    
    date = pd.to_datetime(row['date_occ'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
    date_str = date.strftime('%Y%m%d') if pd.notna(date) else '00000000'
    time_str = str(row['time_occ']).zfill(4)[:4]  # HHMM
    area = str(row['area'])
    crm_cd = str(row['crm_cd'])
    dr_no = str(row['dr_no'])
    
    return f"E_{date_str}_{time_str}_{area}_{crm_cd}_{dr_no}"
```

# Function to push data into Hbase

```python
def push_function(df, table_name, catalog, limit=550000):
    
    connection = happybase.Connection('localhost',port=9090)
    table = connection.table(table_name)
    
    for idx, row in df.head(limit).iterrows():
        rowkey = implement_rowkey(row) 

        data = {}
        for col, meta in catalog.items():
            value = row.get(col) 
            if pd.notna(value):
                family = meta['family']
                qualifier = meta['qualifier']
                data[f"{family}:{qualifier}"] = str(value)

        if data:
          
            table.put(rowkey.encode('utf-8'), data)
    
    print(f"{min(limit, len(df))} rows insert into HBase.")
```

```python
push_function(df, table_name="practice:crimes", catalog=catalog, limit=550000)
```

# Victim age and sex for crimes of INTIMATE PARTNER - SIMPLE ASSAULT 

```python

import happybase
import pandas as pd

connection = happybase.Connection('localhost', port=9090)
table = connection.table('practice:crimes')

rows = table.scan()
result = []

for key, data in rows:
    desc = data.get(b'crime_details:crm_cd_desc', b'').decode('utf-8')
    if desc == 'INTIMATE PARTNER - SIMPLE ASSAULT':
        result.append({
            
            'rowkey': key.decode(),
            'vict_age': data.get(b'crime_details:vict_age', b'').decode('utf-8'),
            'vict_sex': data.get(b'crime_details:vict_sex', b'').decode('utf-8')
        })

print("INTIMATE PARTNER - SIMPLE ASSAULT victims:")
for r in result[:10]:
    print(r)
```

# All SHOPLIFTING and VANDALISM crimes (if the label of the crime contains it) 

```python

import happybase
import pandas as pd

connection = happybase.Connection('localhost', port=9090)
table = connection.table('practice:crimes')


rows = table.scan()
a = []

for key, data in rows:
    desc = data.get(b'crime_details:crm_cd_desc', b'').decode('utf-8').lower()
    if 'shoplifting' in desc or 'vandalism' in desc:
        a.append({
            'rowkey': key.decode(),
            'crm_cd_desc': desc
        })

print("SHOPLIFTING / VANDALISM crimes:")
for r in a[:10]:
    print(r)
```

# Crimes occuring in Hollywood 

```python

import happybase
import pandas as pd

connection = happybase.Connection('localhost', port=9090)
table = connection.table('practice:crimes')


rows = table.scan()
b = []

for key, data in rows:
    area_name = data.get(b'place:area_name', b'').decode('utf-8').upper()
    if area_name in  ['HOLLYWOOD','N HOLLYWOOD'] :
        b.append({
            'rowkey': key.decode(),
            'area_name': area_name
        })

print("Crimes in HOLLYWOOD:")
for r in b[:10]:
    print(r)
```
